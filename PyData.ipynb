SDSS Query
----------
SELECT TOP 1000 S.objID, GN.distance, S.u, S.g, S.r, S.i, S.z, S.err_u, S.err_g, S.err_r, S.err_i, S.err_z
FROM Star as S
JOIN dbo.fGetNearbyObjEq(229.0128,-0.1082, 3) AS GN
ON S.objID = GN.objID
WHERE ((S.flags_r & 0x10000000) != 0)
-- detected in BINNED1
AND ((S.flags_r & 0x8100000c00a4) = 0)
-- not EDGE, NOPROFILE, PEAKCENTER, NOTCHECKED, PSF_FLUX_INTERP,
-- SATURATED, or BAD_COUNTS_ERROR
AND (((S.flags_r & 0x400000000000) = 0) or (S.psfmagerr_r <= 0.2))
-- not DEBLEND_NOPEAK or small PSF error
-- (substitute psfmagerr in other band as appropriate)
AND (((S.flags_r & 0x100000000000) = 0) or (S.flags_r & 0x1000) = 0)
-- not INTERP_CENTER or not COSMIC_RAY
