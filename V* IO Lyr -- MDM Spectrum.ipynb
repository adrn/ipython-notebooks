```python
d1 = np.squeeze(f1[0].data)
d2 = np.squeeze(f2[0].data)
```

```python
figsize(14,8)
plot(d1[1], marker=None, drawstyle='steps')
plot(d2[1], marker=None, drawstyle='steps', color='r')
```

```python
plot(d1[0], marker=None, drawstyle='steps', color='r')
plot(d2[0], marker=None, drawstyle='steps', color='k')
```

```python
def model(p, wvln, flux):
    return flux[1] + p[0]*wvln + p[1]

def err_func(p, wvln1, flux1, flux2):
    return flux2 - model(p, wvln1, flux1)
```

```python
flux1, flux2 = d1[0], d2[0]
ix = np.arange(len(d1[0]))
p_opt, ier = leastsq(err_func, x0=[0., 0.], args=(ix, flux1, flux2))
```

```python
plot(ix, flux1, marker=None, drawstyle='steps', color='k')
plot(ix, model(p_opt, ix, flux2), marker=None, drawstyle='steps', color='k')
#plot(d2[0], marker=None, drawstyle='steps', color='k')
```
