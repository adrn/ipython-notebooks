```python
from skimage import data, io, filter
import Image
import scipy.ndimage as snd
from skimage.morphology import skeletonize
```

```python
filename = "/Users/adrian/Documents/Research/old/p_aeruginosa_colonies/1.tif"
image = np.asarray(Image.open(filename))
image.shape
```

```python
io.imshow(image[...,0])
```

```python
edges = filter.sobel(image[...,0])
edges.shape
big_edges = np.where(edges > (3.*np.median(edges)), edges, 0.)
```

```python
io.imshow(big_edges)
```

```python
smoothed_big_edges = snd.gaussian_filter(big_edges, 2.0)
smoothed_big_edges = np.where(smoothed_big_edges > (3.*np.median(smoothed_big_edges)), 1., 0.)
io.imshow(smoothed_big_edges)
```

```python
figure(figsize=(13,10))
io.imshow(image[...,0])
xx = snd.gaussian_filter(big_edges, 2.)
io.imshow(skeletonize(np.where(xx > (3.*np.median(xx)), 1., 0.)), alpha=0.5)
```
