```python
star_idx = (all_data['class'] == "STAR") # & np.array([("A" in x) or ("F" in x) for x in all_data['subClass']])
star_data = all_data[star_idx]
```

```python
star_coords = coord.ICRSCoordinates(star_data['RA']*u.degree, star_data['Dec']*u.degree)
```

```python
star_sgr_c = star_coords.transform_to(SgrCoordinates)
```

```python
idx = (np.fabs(star_sgr_c.Beta) < 10.*u.degree) & (star_data['mag1'] < 60)
```

```python
sgr_plane_stars = star_data[idx]
sgr_plane_L = star_sgr_c.Lambda.degree[idx]
sgr_plane_B = star_sgr_c.Beta.degree[idx]
```
