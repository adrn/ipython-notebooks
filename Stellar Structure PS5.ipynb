```python
from __future__ import division
import copy
import scipy.optimize as so
import numpy as np

# SI constants
G = 6.67384E-11
m_e = 9.10938291e-31 
m_p = 1.672621777e-27
h = 6.62606957e-34
M_sun = 1.9891e+30
```

# Problem 1

The lookback time is given by

\begin{equation}
t_L = t_H\int^z_0 \frac{dz'}{(1 + z')E(z')}
\end{equation}

where 
\begin{equation}
    E(z') = \sqrt{\Omega_M(1+z)^3 + \Omega_k(1+z)^2 + \Omega_\lambda}
\end{equation}

and
\begin{equation}
    t_H = 9.78\times10^9 h^{-1} yr
\end{equation}


```python
cosmo_kwargs = {"omega_M" : 0.214, "omega_k" : 0, "omega_L" : 0.742}

def E(z, omega_M, omega_k, omega_L):
    return np.sqrt(omega_M*(1+z)**3 + omega_k*(1+z)**2 + omega_L)

def t_H(h):
    return 9.78 / h # Gigayears

def integrand(z):
    return 1./((1 + z)*E(z, **cosmo_kwargs))
```

The Extended Simpson's rule is written:

\begin{equation}
\int^{x_{2n}}_{x_0} f(x) dx = \frac{1}{3}h [f_0 + 4(f_1+f_3+...+f_{2n-1}) + 2(f_2+f_4+...+f_{2n-2}) + f_{2n}]
\end{equation}

```python
def extended_simpson(f, a, b, hh):
    ''' Integrate the function f from a to b using the extended Simpson's rule with a step size of h '''
    grid = np.arange(a, b, hh)
    
    sum = f(a)
    for ii, z_i in enumerate(grid):
        if ii == 0:
            continue
        elif ii % 2 != 0:
            sum += 4.*f(z_i)
        elif ii % 2 == 0:
            sum += 2.*f(z_i)
            
    sum += f(b)

    return (hh/3)*sum
```

```python
z1,z2 = (0., 10.)
step_sizes = (z2-z1)/np.arange(10000, 10, -100)
lookback_times = t_H(0.719)*np.array([extended_simpson(integrand, z1, z2, step_size) for step_size in step_sizes])
```

```python
figure(figsize=(15,8))
semilogx(step_sizes, lookback_times)
xlabel("Step size")
ylabel("Lookback time")
```

```python
best_lookback_time = t_H(0.719) * extended_simpson(integrand, 0., 10., 0.001)
error = np.fabs(lookback_times-best_lookback_time)
p,ier = so.leastsq(lambda p,x,y: y-p[0]*x**p[1]-p[2], [1.,1.,1.], args=(step_sizes, error))
p, ier
```

```python
figure(figsize=(10,8))
loglog(step_sizes, error)
loglog(step_sizes, p[0]*step_sizes**p[1]+p[2], label=r"$\propto h^4$")
xlabel(r"Step size, $h$")
ylabel("Relative error")
legend()
```

```python
def monte_carlo_pi(N):
    ''' Compute the area enclosed in the unit circle (e.g. π) using Monte-Carlo integration '''
    num_inside = 0
    x = np.random.uniform(size=int(N))
    y = np.random.uniform(size=int(N))
    num_inside = sum((x**2 + y**2) < 1)
    
    return 4.*num_inside / float(N)
            
```

```python
Ns = np.logspace(0, 7, 100)
errors = []
for N in Ns:
    errors.append(abs(np.pi - monte_carlo_pi(N)))
```

```python
figure(figsize=(10,8))
xscale("log")
yscale("log")
scatter(Ns, errors, color='k')
plot(Ns, Ns**(-1/2.), 'r--', label=r"$N^{-1/2}$")
xlabel("N samples")
ylabel("Error")
legend()
```

# Problem 3

The differential equations we are trying to solve are (worked out on attached sheet of paper):

\begin{equation}
\frac{d\rho}{dr} = -\frac{G\rho^{1/3}m}{r^2}12\big(\frac{\pi}{3}\big)^{2/3}\frac{m_e(2m_p)^{5/3}}{h^2} = K \frac{m\rho^{1/3}}{r^2}
\end{equation}

\begin{equation}
\frac{dm}{dr} = 4\pi r^2 \rho
\end{equation}

```python
class RK4Integrator(object):

    def __init__(self, *args):
        ''' Initialize with a parameter list of functions that represent
            the coupled ODE
        '''
        if len(args) == 0:
            raise ValueError("You must pass at least function in to be integrated!")
        self.functions = list(args)

    def initialize(self, *args):
        ''' Initial positions for all variables. If there are N coupled functions,
            you must specify N initial conditions.
        '''
        if len(args) != len(self.functions):
            raise ValueError("If you passed in N functions, you must pass N initial conditions.")
        self.initial_conditions = list(args)

    def run(self, x1, x2, step_size=None):
        ''' Integrate the functions from x1 to x2 with the given step size. If step_size is
            None, will use an adaptive step size.
        '''
        container = []
        self.params = copy.copy(self.initial_conditions)
        xn = x1
        container.append([xn] + self.params)

        if step_size == None:
            # Adaptive step size
            raise NotImplementedError()

        else:
            # Fixed step size
            while xn < x2:
                try:
                    xn, new_params = self._step(xn, step_size)
                    container.append([xn] + new_params)
                    self.params = new_params

                except ValueError: # density goes negative, can't take 1/3 power of it so it breaks
                    break

        return np.array(container).T

    def _step(self, xn, step_size):
        ''' Take one step of the given step size from xn. '''

        coeff_container = dict(a=[],b=[],c=[],d=[])
        for func in self.functions:
            coeff_container["a"].append(func(xn, *self.params)*step_size)

        # Just updated values of the parameters used for computation -- not the final "new" params
        mid_params = [xx + coeff_container["a"][ii]/2. for ii,xx in enumerate(self.params)]
        for func in self.functions:
            coeff_container["b"].append(func(xn+step_size/2., *mid_params)*step_size)

        mid_params = [xx + coeff_container["b"][ii]/2. for ii,xx in enumerate(self.params)]
        for func in self.functions:
            coeff_container["c"].append(func(xn+step_size/2., *mid_params)*step_size)

        mid_params = [xx + coeff_container["c"][ii]/2. for ii,xx in enumerate(self.params)]
        for func in self.functions:
            coeff_container["d"].append(func(xn+step_size, *mid_params)*step_size)

        xn += step_size

        new_params = [0. for xx in self.params]
        for ii in range(len(self.params)):
            new_params[ii] = self.params[ii] + (1/6.) * (coeff_container["a"][ii] + 2.*coeff_container["b"][ii] + 2.*coeff_container["c"][ii] + coeff_container["d"][ii])

        return xn, new_params
```

```python
constant_coeff = G * 12 * (np.pi/3)**(2./3) * m_e/h**2. * (2*m_p)**(5./3)
print constant_coeff
```

```python
def F_rho(r, rho, m):
    return -constant_coeff * rho**(1./3) * m / r**2 

def F_m(r, rho, m):
    return 4*np.pi*r**2 * rho
```

```python
initial_r = 1.
rho_c = 1E8
integrator = RK4Integrator(F_rho, F_m)
integrator.initialize(rho_c, 4./3*np.pi*initial_r**3*rho_c)
data = integrator.run(initial_r, 1E8, step_size=1E5) # I chose 10^5 m because that is around 1/100 of the radius of a massive white dwarf
```

```python
fig, axes = subplots(2, 1, sharex=True, figsize=(12,16))
axes[0].loglog(data[0], data[1])
axes[0].set_ylim(1E6, 2E8)
axes[0].set_ylabel(r"Density ($\mathrm{kg}$ $\mathrm{m}^{-3}$)")

axes[1].loglog(data[0], data[2])
axes[1].set_xlabel("Radius (m)")
axes[1].set_ylabel("Mass (kg)")
fig.subplots_adjust(hspace=0.0)
```

```python
initial_r = 1.

max_radii = []
total_masses = []

fig, axes = subplots(2, 1, sharex=True, figsize=(12,16))
for rho_c in np.logspace(6,10,20):
    integrator.initialize(rho_c, 4./3*np.pi*initial_r**3*rho_c)
    r, density, mass = integrator.run(initial_r, 1E9, step_size=1E5)
    axes[0].loglog(r, density)
    axes[1].loglog(r, mass)
    
    max_radii.append(max(r[np.isfinite(mass)]))
    total_masses.append(max(mass[np.isfinite(mass)]))

axes[0].set_title("Density and Mass vs. Radius for a variety of central densities")
axes[0].set_ylim(1E4, 2E10)
axes[0].set_ylabel(r"Density ($\mathrm{kg}$ $\mathrm{m}^{-3}$)")

axes[1].set_xlabel("Radius (m)")
axes[1].set_ylabel("Mass (kg)")
fig.subplots_adjust(hspace=0.0)
```

```python
p2,ier = so.leastsq(lambda p,x,y: y-p[0]*x-p[1], [1.,1.], args=(np.log10(max_radii), np.log10(total_masses)))
p2, ier
```

```python
figure(figsize=(12,8))
title("Mass Radius relationship for White Dwarfs")
loglog(max_radii, total_masses, linestyle='none', marker='o', label="Data")
loglog(max_radii, (10**p2[1])*max_radii**p2[0], color='r', linestyle='--', linewidth=3., label=r"Fit for $M \propto R^{-3}$")
xlim(5E6, 5E7)
legend()
xlabel("Outer radius")
ylabel("Total Mass")
```

```python
def F_rho2(r, rho, m):
    return -1.02415E-20 * rho**(1./3) * m / r**2 * np.sqrt(rho**(2./3) + 1.567250E6)

def F_m2(r, rho, m):
    return 4*np.pi*r**2 * rho
```

```python
initial_r2 = 1E3
integrator2 = RK4Integrator(F_rho2, F_m2)

max_radii = []
total_masses = []
fig, axes = subplots(2, 1, sharex=True, figsize=(12,16))
for rho_c2 in np.logspace(8, 13, 20):
    integrator2.initialize(rho_c2, 4.*np.pi*initial_r2**3*rho_c2)
    r, density, mass = integrator2.run(initial_r2, 1E9, step_size=1E5)
    axes[0].loglog(r, density)
    axes[1].loglog(r, mass)
    
    max_radii.append(max(r[np.isfinite(mass)]))
    total_masses.append(max(mass[np.isfinite(mass)]))

axes[0].set_title("Density and Mass vs. Radius for a variety of central densities. Note the maximum allowed mass.")
axes[0].set_ylim(1E4, 2E13)
axes[0].set_ylabel(r"Density ($\mathrm{kg}$ $\mathrm{m}^{-3}$)")

axes[1].axhline(max(total_masses), label=r"Max. allowed mass = {0:.2f} $M_\odot$".format(max(total_masses)/M_sun), color='r', linestyle='--', linewidth=2.)
axes[1].set_xlabel("Radius (m)")
axes[1].set_ylabel("Mass (kg)")
axes[1].legend(loc="lower right")
fig.subplots_adjust(hspace=0.0)
```
