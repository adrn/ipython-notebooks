```python
def gaussian2D(p, x, y):
    A, mu_x, mu_y, sigma_x, sigma_y, rho_xy = p
    cov = np.array([[sigma_x**2, rho_xy*sigma_x*sigma_y],
                    [rho_xy*sigma_x*sigma_y, sigma_y**2]])
    
    xx = np.array([x,y])
    mu = np.array([mu_x, mu_y])[:,np.newaxis]
    inv_var = np.linalg.inv(cov)
    
    quad_form = np.diag(np.inner((xx-mu).T, np.inner((xx-mu).T, inv_var)))

    f = A/np.sqrt(2*np.pi)/np.linalg.det(cov)
    return f * np.exp(-0.5 * quad_form)

def errfunc(p, x, y, z):
    return z - gaussian2D(p, x, y)
```
