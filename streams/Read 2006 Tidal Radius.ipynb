```python
from astropy.constants import G
from scipy.signal import argrelmin, argrelmax

from streams import usys
from streams.potential.lm10 import LawMajewski2010
from streams.io import SgrSimulation
from streams.integrate import LeapfrogIntegrator
```

```python
mass = "2.5e8"
m = float(mass)
potential = LawMajewski2010()
sim = SgrSimulation(mass)

p_bound = sim.particles(N=0, expr="tub == 0", tail_bit=False)
sat_vdisp = np.sqrt(np.sum(np.var(p_bound._X[:,3:], axis=0)))
```

```python
s = sim.satellite()
integrator = LeapfrogIntegrator(potential._acceleration_at,
                                np.array(s._X[...,:3]), np.array(s._X[...,3:]),
                                args=(1, np.zeros_like(s._X[...,:3])))
ts, rs, vs = integrator.run(t1=sim.t1, t2=sim.t2, dt=-1.)
s_orbit = np.vstack((rs[:,0][:,np.newaxis].T, vs[:,0][:,np.newaxis].T)).T
```

```python
R = np.squeeze(np.sqrt(np.sum(s_orbit[...,:3]**2, axis=-1)))
V = np.squeeze(np.sqrt(np.sum(s_orbit[...,3:]**2, axis=-1)))

gal_vdisp = potential._parameter_dict['v_halo']
r_tide = np.squeeze(potential._tidal_radius(m, s_orbit[...,:3]))
#sat_vdisp = V * r_tide / R
```

```python
def read_rtide(R, sat_vdisp, gal_vdisp, alpha=1.):
    peris = np.median(R[argrelmin(R)])
    apos = np.median(R[argrelmax(R)])
    L = (2 * apos**2 * peris**2 / (peris**2 - apos**2) * np.log(peris/apos))
    return np.sqrt(L*sat_vdisp**2/gal_vdisp**2) * (-alpha + np.sqrt(alpha**2 + 1 + R**2/L)) / (1 + L/R**2)
```

```python
rts = np.array([read_rtide(R, sat_vdisp, gal_vdisp, alpha) for alpha in (-1,0,1)])
r_tide = np.squeeze(r_tide)
```
