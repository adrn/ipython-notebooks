```python
import triangle
import numpy as np
from sklearn import svm
```

```python
training_data = np.load("/Users/adrian/Downloads/training_data.npy")
training_data = np.vstack(training_data)
all_data = np.load("/Users/adrian/Downloads/all_data.npy")

Ntrain = len(training_data)
Ndata = len(all_data)
```

```python
for ii in range(1):
    idx1 = np.zeros(Ntrain).astype(bool)
    idx1[np.random.randint(Ntrain, size=Ntrain//2)] = True
    
    idx2 = np.zeros(Ndata).astype(bool)
    idx2[np.random.randint(Ndata, size=Ndata//2)] = True
    
    train_X = training_data[idx1]
    train_Y = np.ones(len(train_X))
    X = all_data[idx2]
    Y = np.zeros(len(X))
    
    train_X = np.vstack((train_X, X))
    train_Y = np.append(train_Y, Y)
    
    rbf_svc = svm.SVC(kernel='rbf', gamma=0.7).fit(train_X, train_Y)
    print np.sum(rbf_svc.predict(training_data[~idx1])) / len(training_data[~idx1])
    print np.sum(rbf_svc.predict(all_data[~idx2])) / len(all_data[~idx2])
```

```python
y = rbf_svc.predict(all_data)
```

```python
idx = y == 1
```
