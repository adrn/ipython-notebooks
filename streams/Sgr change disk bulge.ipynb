```python
import astropy.units as u
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from streamteam.integrate import LeapfrogIntegrator
from streams.potential import MiyamotoNagaiPotential, HernquistPotential, LogarithmicPotentialLJ, CompositePotential
```

```python
usys = (u.kpc, u.M_sun, u.Myr, u.radian)

def get_orbit(m_disk=1.E11*u.M_sun, a=6.5*u.kpc, b=0.26*u.kpc, 
              m_bulge=3.4E10*u.M_sun, c=0.7*u.kpc):

    potential = CompositePotential(units=usys)
    potential["disk"] = MiyamotoNagaiPotential(units=usys,
                                       m=m_disk, a=a, b=b)

    potential["bulge"] = HernquistPotential(units=usys,
                                   m=m_bulge, c=c)

    potential["halo"] = LogarithmicPotentialLJ(units=usys,
                                       q1=1.38,
                                       q2=1.,
                                       qz=1.36,
                                       phi=97.*u.degree,
                                       v_halo=121.858*u.km/u.s,
                                       R_halo=12.*u.kpc)
    
    def acceleration(r):
        return potential.acceleration_at(r*u.kpc).value
    
    sgr_w = np.array([19.0,2.7,-6.9,0.2352238,-0.03579493,0.19942887]) # kpc and kpc/Myr
    
    dt = -1.
    t1,t2 = 6200., 0.
    integrator = LeapfrogIntegrator(acceleration)
    ts, qs, ps = integrator.run(q_i=sgr_w[:3], p_i=sgr_w[3:],
                                t1=t1, t2=t2, dt=dt)
    
    return ts, qs, ps
```

```python
plt.figure(figsize=(5,5))

ts,qs,ps = get_orbit()
r = np.sqrt(np.sum(qs[:,0]**2, axis=-1))
print r.min(), r.max()
plt.plot(qs[:,0,0], qs[:,0,2], marker=None)

ts,qs,ps = get_orbit(a=10*u.kpc)
r = np.sqrt(np.sum(qs[:,0]**2, axis=-1))
print r.min(), r.max()
plt.plot(qs[:,0,0], qs[:,0,2], marker=None)

ts,qs,ps = get_orbit(b=0.5*u.kpc)
r = np.sqrt(np.sum(qs[:,0]**2, axis=-1))
print r.min(), r.max()
plt.plot(qs[:,0,0], qs[:,0,2], marker=None)
plt.xlim(-60,60)
plt.ylim(-60,60)
```
