```python
d = io.read_hdf5("/Users/adrian/projects/streams/data/observed_particles/2.5e{}_exp4.hdf5".format(_m))
gc_particles = d['particles'].to_frame(galactocentric)
```

```python
N = 32

D = d['true_particles']._X[:,2]
x,y,z,vx,vy,vz = gc_particles._X.T

print (x[:10])
```

```python
fig,axes = plt.subplots(1,2,figsize=(16,8))
fig2,axes2 = plt.subplots(1,2,figsize=(16,8))
axes[0].plot(x,z,marker='.',linestyle='none',alpha=0.5)
axes[1].plot(vx,vz,marker='.',linestyle='none',alpha=0.5)

axes2[0].plot(x,y,marker='.',linestyle='none',alpha=0.5)
axes2[1].plot(vx,vy,marker='.',linestyle='none',alpha=0.5)

leading = []
trailing = []
lead_c = 'g'
trail_c = 'r'
while True:
    ix = np.random.randint(len(x))
    if np.isnan(d['true_particles'].beta[ix]):
        continue
    
    #print(d['true_particles'].tail_bit[ix],ix)
    if d['true_particles'].beta[ix] < 0:
        leading.append(ix)
        c = lead_c
    else:
        trailing.append(ix)
        c = trail_c
    
    axes[0].plot(x[ix], z[ix], label="tub={}".format(d['true_particles'].tub[ix]),
                 marker='o',linestyle='none',alpha=0.85,ms=10, color=c)
    axes[1].plot(vx[ix], vz[ix], label="tub={}".format(d['true_particles'].tub[ix]),
                 marker='o',linestyle='none',alpha=0.85,ms=10, color=c)
    
    axes2[0].plot(x[ix], y[ix], label="tub={}".format(d['true_particles'].tub[ix]),
                 marker='o',linestyle='none',alpha=0.85,ms=10, color=c)
    axes2[1].plot(vx[ix], vy[ix], label="tub={}".format(d['true_particles'].tub[ix]),
                 marker='o',linestyle='none',alpha=0.85,ms=10, color=c)
    
    if len(leading) >= N and len(trailing) >= N:
        break
    
# now plot observed positions
# x,y,z,vx,vy,vz = d['particles'].to_frame(galactocentric)._X.T
# for ix in idx:
#     axes[0].plot(x[ix], z[ix], marker='o', linestyle='none',alpha=0.85,ms=6)
#     axes[1].plot(vx[ix], vz[ix], marker='o',linestyle='none',alpha=0.85,ms=6)
    
#legend(loc='lower left')
```
