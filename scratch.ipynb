```python
def ln_gaussian(x, mu, cov):
    xx = x - mu
    inv_cov = np.linalg.inv(cov)
    exponent = xx.dot(np.dot(inv_cov, xx.T))
    A = -0.5*np.log(2*np.pi*np.linalg.det(cov))
    
    return A - 0.5*np.diagonal(exponent)
```

```python
def test_ln_gaussian():
    x,y = np.meshgrid(np.linspace(0.,10,100), np.linspace(0.,10,100))
    x,y = x.ravel(), y.ravel()
    cov = np.array([[1.,1.2],
                    [1.2,2.]])**2.
    g = np.exp(ln_gaussian(np.vstack((x,y)).T, np.array([4.,5.]), cov))
    g = g.reshape(100,100)
    imshow(g)
    
test_ln_gaussian()
```

```python
def gaussian_exp(x, mu, cov):
    xx = x - mu
    inv_cov = np.linalg.inv(cov)
    exponent = xx.dot(np.dot(inv_cov, xx.T))
    A = (2*np.pi*np.linalg.det(cov))**-0.5
    B = 0.5*np.diagonal(exponent)
    return B*np.log(A)
```

```python
def test_gaussian_exp():
    x,y = np.meshgrid(np.linspace(0.,10,100), np.linspace(0.,10,100))
    x,y = x.ravel(), y.ravel()
    cov = np.array([[1.,1.2],
                    [1.2,2.]])**2.
    g = np.exp(gaussian_exp(np.vstack((x,y)).T, np.array([4.,5.]), cov))
    g = g.reshape(100,100)
    imshow(g)
    
test_gaussian_exp()
```

```python
def model(p_orbits, s_orbit, r_tide, v_esc, t_idx):
    R,V = np.squeeze(r_tide[t_idx]), np.squeeze(v_esc[t_idx])
    V_diag = np.array([R/np.sqrt(3.)]*3 + [V/np.sqrt(3.)]*3)
    V = np.diag(V_diag)
    
    #x_s = np.append(p_orbits._r[t_idx, ii], p_orbits._v[t_idx, ii])
    x_s = np.hstack((p_orbits._r[t_idx], p_orbits._v[t_idx]))
    x_p = np.append(s_orbit._r[t_idx], s_orbit._v[t_idx])
    
    return gaussian_exp(x_s, x_p, V)
```

```python
def ln_posterior(p, *args):
    param_names, particles, satellite, t1, t2, resolution = args
    return ln_prior(p, param_names) + ln_likelihood(p, *args)
```

```python
def ln_likelihood(halo_p, halo_p_names, particles, satellite, t1, t2, resolution):
    """ Evaluate the likelihood function for a given set of halo 
        parameters.
    """
    halo_params = dict(zip(halo_p_names, halo_p))
    
    # LawMajewski2010 contains a disk, bulge, and logarithmic halo 
    lm10 = LawMajewski2010(**halo_params)
    
    integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
    
    # not adaptive: s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
    s_orbit,p_orbits = integrator.run(timestep_func=timestep,
                                      timestep_args=(lm10, satellite.m.value),
                                      resolution=resolution,
                                      t1=t1, t2=t2)
    
    r_tide = lm10._tidal_radius(m=s_orbit._m,
                                r=s_orbit._r)[:,:,np.newaxis]
    v_esc = lm10._escape_velocity(m=s_orbit._m,
                                  r_tide=r_tide)
    
    ys = np.zeros((len(particles),len(r_tide)))
    for jj in range(len(r_tide)):
        ys[:,jj] = model(p_orbits, s_orbit, r_tide, v_esc, jj) / len(r_tide)
    
    print np.exp(ys)
    return
    
    #dts = np.fabs(s_orbit._t[1:]-s_orbit._t[:-1])
    #integral = np.sum(logsumexp(ys[:,1:]*np.log(dts), axis=1))
    #return integral

    integral = integrate.simps(np.exp(ys), x=s_orbit._t)
    #integral = integrate.cumtrapz(np.exp(ys), x=s_orbit._t)
    return np.sum(np.log(integral))
```
