```python
import ptf.analyze as pa
import ptf.lightcurve as pl
import numpy as np
import matplotlib.pyplot as plt
import copy
```

The first thing I need to do is generate a set of light curves that just have Gaussian noise. 

```python
def generate_base_lightcurves(N_light_curves=1000, baseline=600.):
    light_curves =[]
    for ii in range(N_light_curves):
        N_data_points = int(10**np.random.normal(1.7, 0.4))
        if N_data_points < 10: continue
        mjd = sorted(np.random.uniform(0., baseline, size=N_data_points))
        light_curves.append(pl.SimulatedLightCurve(mjd, mag=15., error=np.fabs(np.random.normal(np.random.uniform(0.01, 0.2), 0.05, size=N_data_points))))
    
    return light_curves
        
```

```python
base_light_curves = generate_base_lightcurves(100)
```

Next, I'll write a routine that:

 * makes two copies of the base light curves
 * adds simulated microlensing events to one copy
 * adds simulated "bad data" to the other

```python
def add_microlensing_events(base_light_curves):
    lcs = copy.deepcopy(base_light_curves)
    for lc in lcs:
        lc.add_microlensing_event()
    return lcs

def add_bad_data(base_light_curves):
    lcs = copy.deepcopy(base_light_curves)
    for lc in lcs:
        u0 = np.random.uniform(0.001, 0.1)
        t0 = lc.mjd[np.random.randint(len(lc))]
        lc.add_microlensing_event(u0=u0, t0=t0)
        np.random.shuffle(lc.mag)
    return lcs
```

```python
ml_light_curves = add_microlensing_events(base_light_curves)
bad_light_curves = add_bad_data(base_light_curves)
```

Now I'll compute all variability indices for each set of light curves, and plot projections of the feature space.

```python
indices = ["j", "k", "delta_chi_squared", "eta", "corr", "con"]
all_indices = ["j", "k", "delta_chi_squared", "eta", "corr", "con", "N", "med_err", "max_err"]
```

```python
ml_var_indices = []
for lc in ml_light_curves:
    try:
        ml_var_indices.append(pa.compute_variability_indices(lc, indices=indices, return_tuple=True) + (len(lc), np.median(lc.error), np.max(lc.error)))
    except ValueError:
        continue
        
bad_var_indices = []
for lc in bad_light_curves:
    try:
        bad_var_indices.append(pa.compute_variability_indices(lc, indices=indices, return_tuple=True) + (len(lc), np.median(lc.error), np.max(lc.error)))
    except ValueError:
        continue
```

```python
ml_var_indices = np.array(ml_var_indices, dtype=[(n, float) for n in all_indices])
bad_var_indices = np.array(bad_var_indices, dtype=[(n, float) for n in all_indices])
```

```python
def ind_vs_ind(x_index, y_index):
    figure()
    scatter(ml_var_indices[x_index], ml_var_indices[y_index], color="r", alpha=0.3)
    scatter(bad_var_indices[x_index], bad_var_indices[y_index], color="b", alpha=0.3)
    xlabel(x_index)
    ylabel(y_index)
    #xscale("symlog")
    #yscale("symlog")
```

```python
for ii,x_index in enumerate(all_indices):
    for jj,y_index in enumerate(all_indices):
        if ii <= jj: continue 
        ind_vs_ind(x_index, y_index)
```

```python
all_var_indices = np.hstack((ml_var_indices, bad_var_indices)).view(np.float64).reshape(all_var_indices.shape + (-1,))
```
