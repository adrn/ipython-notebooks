1) Fit a spline (when possible) to the peak, following the contours of
the lightcurve bump as closely as possible, and find the time of
maximum brightness and the times, before and after the t_max, at which
the lightcurve returns to baseline (t_base_1, t_base_2).

2) Choose a sequence of brightness intervals between baseline and peak.

3) For each brightness interval, using the spline, calculate the time
before and the time after the peak at which the lightcurve crosses the
brightness threshold (t_before, t_after).  Then calculate the
difference from the t_max:
dt_before = t_max - t_before
dt_after = t_after - t_max

4) Plot dt_before, after as functions of brightness.  If the peak is
symmetrical, the curves should lie on top of one another.

```python
import numpy as np
import scipy.optimize as so
import scipy.interpolate as si
from ptf.lightcurve import PTFLightCurve
```

```python
t = np.linspace(0, 20, 50)
sigma = 0.5
mag = t * (1/np.sqrt(2*np.pi*sigma)) * np.exp( - (t-10)**2 / (2*sigma**2))
plot(t, mag, color="k", marker=".", linestyle="none")
xlim(0,20)
ylim(-1, 10)
```

```python
tck = si.splrep(t,mag,s=0)

t_new = np.linspace(0,20,200)
mag_new = si.splev(t_new,tck,der=0)

plot(t, mag, color="k", marker=".", linestyle="none")
plot(t_new, mag_new, color="r", marker=".", linestyle="none")
xlim(0,20)
ylim(-1, 10)
```
