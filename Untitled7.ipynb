```python
def sigma(t):
    A = 0.5
    w = 1.
    return A*np.cos(w*t)**2 + 0.1

def mu(t):
    A = 0.5
    w = 1.
    return A*np.cos(w*t)

def gaussian(x, t):
    return 1./np.sqrt(2*np.pi)/sigma(t) * np.exp(-0.5 * (x-mu(t))**2 / sigma(t)**2)

x = np.linspace(-3, 3, 100)
t = np.linspace(0, 10, 100)
X,T = np.meshgrid(x,t)

g = gaussian(X.ravel(), T.ravel()).reshape(X.shape)
```
