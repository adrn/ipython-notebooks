```python
import numpy as np
from scipy.stats import norm
from IPython.html import widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,250)
def special_sin(a):
    y = np.sin(a*x)
    plt.figure(figsize=(6,6))
    plt.plot(x, y, marker=None, linestyle='-')
    plt.xlim(-10,10)
    plt.show()
```

```python
a = widgets.FloatSliderWidget(min=0.1, max=5., step=0.05, value=0.1)
widgets.interact(special_sin, a=a)
```
