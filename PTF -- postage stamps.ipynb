```python
import os
import aplpy
from astropy.io import fits
import astropy.coordinates as coord
import astropy.units as u
from skimage.data import imread
```

```python
def height_above_plane(ra, dec, D):
    icrs = coord.ICRSCoordinates(ra, dec, unit=(u.degree,u.degree))
    g = icrs.galactic
    return D*np.sin(g.b.radians)
```

```python
d1 = height_above_plane(94.5010416, 20.528472, 0.8)
d2 = height_above_plane(94.5010416, 20.528472, 1.1)
print "PTFS1206i: {0} to {1} kpc".format(d1,d2)

d1 = height_above_plane(243.76014, 54.01497, 0.7)
d2 = height_above_plane(243.76014, 54.01497, 1.3)
print "PTFS1216bt: {0} to {1} kpc".format(d1,d2)

d1 = height_above_plane(262.1086, 69.417, 1.9)
d2 = height_above_plane(262.1086, 69.417, 4.2)
print "PTFS1217cv: {0} to {1} kpc".format(d1,d2)
```
