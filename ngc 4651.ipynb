```python
fig = figure(figsize=(22,24))
gs = gridspec.GridSpec(4, 3)

# Halpha
ax = fig.add_subplot(gs[0,0])
ax.plot(wave/(1+spec_hdulist[2].data["Z"]), flux, 'k-')
ax.axvline(6564.6, color='r', linestyle="--")
ax.annotate(r"$H_\alpha$", (6564.6, 325), fontsize=24)
ax.set_xlim(6540, 6595)
[xtick.set_visible(False) for xtick in ax.get_xticklabels()]

# Hbeta
ax = fig.add_subplot(gs[0,1])
ax.plot(wave/(1+spec_hdulist[2].data["Z"]), flux, 'k-')
ax.axvline(4862.7, color='r', linestyle="--")
ax.annotate(r"$H_\beta$", (4862.7, 325), fontsize=24)
ax.set_xlim(4840, 4880)
[ytick.set_visible(False) for ytick in ax.get_yticklabels()]
[xtick.set_visible(False) for xtick in ax.get_xticklabels()]

# Hgamma
ax = fig.add_subplot(gs[0,2])
ax.plot(wave/(1+spec_hdulist[2].data["Z"]), flux, 'k-')
ax.axvline(4341.7, color='r', linestyle="--")
ax.annotate(r"$H_\gamma$", (4341.7, 325), fontsize=24)
ax.set_xlim(4320, 4360)
[ytick.set_visible(False) for ytick in ax.get_yticklabels()]
[xtick.set_visible(False) for xtick in ax.get_xticklabels()]

# Ca H,K
ax = fig.add_subplot(gs[3,0])
ax.plot(wave/(1+spec_hdulist[2].data["Z"]), flux, 'k-')
ax.axvline(3933.7, color='r', linestyle="--")
ax.axvline(3968.9, color='r', linestyle="--")
ax.annotate(r"$Ca$ $H$,$K$", (3933., 55), fontsize=24)
ax.set_xlim(3910, 3990)
ax.set_ylim(15, 60)
[xtick.set_visible(False) for xtick in ax.get_xticklabels()]

# CaT
ax = fig.add_subplot(gs[3,1])
ax.plot(wave/(1+spec_hdulist[2].data["Z"]), flux, 'k-')
ax.axvline(8500.7, color='r', linestyle="--")
ax.axvline(8544.9, color='r', linestyle="--")
ax.axvline(8664.9, color='r', linestyle="--")
ax.annotate(r"$CaII$ triplet", (8545., 42), fontsize=24)
ax.set_xlim(8480, 8680)
ax.set_ylim(15, 60)
[xtick.set_visible(False) for xtick in ax.get_xticklabels()]

# NaII doublet
ax = fig.add_subplot(gs[3,2])
ax.plot(wave/(1+spec_hdulist[2].data["Z"]), flux, 'k-')
ax.axvline(5891.95, color='r', linestyle="--")
ax.axvline(5897.12, color='r', linestyle="--")
ax.annotate(r"$NaI$ doublet", (5889., 52), fontsize=24)
ax.set_xlim(5870, 5920)
ax.set_ylim(15, 60)
[xtick.set_visible(False) for xtick in ax.get_xticklabels()]

# Whole spectrum
ax = fig.add_subplot(gs[1:3,:])
ax.semilogy(wave/(1+spec_hdulist[2].data["Z"]), flux, "k-")
ax.set_xlim(wave.min(), wave.max())
ax.set_ylim(flux.min()-1., flux.max()+10.)

fig.subplots_adjust(wspace=0.05, hspace=0.05)
```

Spectral Features
=================
 * The strong Balmer emission lines ($H_\alpha$, $H_\beta$, $H_\gamma$) are indicative of star formation
 * $CaII$ absorption is a proxy for metallicity - some older population

```python
def zoom_line(wvln):
    min_w, max_w = wvln-100, wvln+100
    
    new_wave = wave/(1+spec_hdulist[2].data["Z"])
    idx = (new_wave > min_w) & (new_wave < max_w)
    new_wave = new_wave[idx]
    new_flux = flux[idx]
    
    figure()
    plot(new_wave, new_flux, 'k-')
    axvline(wvln, color='r', linestyle="--")
    xlim(wvln-100, wvln+100)
```

```python
zoom_line(5200)
```
