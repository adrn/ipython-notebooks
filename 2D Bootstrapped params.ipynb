```python
import os
import cPickle as pickle
import astropy.units as u
from streams.potential.lm10 import halo_params

plot_path = "/Users/adrian/projects/streams/plots/hotfoot/"
```

```python
def read_data(subdir):
    fpath = os.path.join(plot_path, subdir, "all_best_parameters.pickle")
    with open(fpath) as f:
        data = pickle.load(f)
    return data
```

```python
def plot_2params(ax, d, d_err, p1, p2):
    ax.scatter(d[p1], d[p2], color='k', marker='.', alpha=0.5)
    ax.scatter(d_err[p1], d_err[p2], color='#666666', marker='^', alpha=0.5)
    ax.set_xlim(lims[p1])
    ax.set_ylim(lims[p2])
    ax.axvline(params[p1], linestyle="--", color="#555555")
    ax.axhline(params[p2], linestyle="--", color="#555555")

def make_scatter_plot(d, d_err, subtitle=""):
    fig,axes = plt.subplots(3,3,figsize=(12,12))
    
    p = ["q1", "qz", "phi", "v_halo"]
    
    params = dict(q1=halo_params["q1"],
                  qz=halo_params["qz"],
                  phi=halo_params["phi"].value,
                  v_halo=halo_params["v_halo"].to(u.km/u.s).value)
    
    lims = dict(q1=(1.2,1.6), qz=(1.2,1.6), phi=(np.pi/2, 1.9), v_halo=(115, 130))
    labels = dict(q1="$q_1$", qz="$q_z$", 
                  phi="$\phi$ [rad]", v_halo="$v_{halo}$ [km/s]")

    plot_2params(axes[0,0], d, d_err, "q1", "qz")
    axes[0,0].set_ylabel(labels["qz"])
    plot_2params(axes[1,0], d, d_err, "q1", "phi")
    axes[1,0].set_ylabel(labels["phi"])
    plot_2params(axes[2,0], d, d_err, "q1", "v_halo")
    axes[2,0].set_xlabel(labels["q1"])
    axes[2,0].set_ylabel(labels["v_halo"])
    
    plot_2params(axes[1,1], d, d_err, "qz", "phi")
    plot_2params(axes[2,1], d, d_err, "qz", "v_halo")
    axes[2,1].set_xlabel(labels["qz"])
    plot_2params(axes[2,2], d, d_err, "phi", "v_halo")
    axes[2,2].set_xlabel(labels["phi"])

    axes[0,1].set_visible(False)
    axes[0,2].set_visible(False)
    axes[1,2].set_visible(False)
    
    axes[0,0].set_xticklabels([])
    axes[1,0].set_xticklabels([])
    axes[1,1].set_xticklabels([])
    
    axes[1,1].set_yticklabels([])
    axes[2,1].set_yticklabels([])
    axes[2,2].set_yticklabels([])
    
    fig.suptitle("Inferred halo parameters for 100 bootstrapped realizations", fontsize=22)
    figtext(.5,.93, subtitle, fontsize=18, ha='center')
    fig.subplots_adjust(hspace=0.04, wspace=0.04)
```
