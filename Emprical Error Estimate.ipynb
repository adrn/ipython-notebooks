```python
from IPython.display import display, Math, Latex
from scipy import optimize
import emcee
```

```python
def err_f(model):
    return lambda p,x,y_data,y_err: (y_data - model(p, x))/y_err

def linear_model(p, x):
    return p[0]*x + p[1]
        
def quadratic_model(p, x):
    return p[0]*x**2 + p[1]*x + p[2]
```

```python
a,b,c = (0.5, 3, 1.)
num_points = 25

x = np.random.random(size=num_points)*10. + 15
x.sort()

ysigma = np.random.uniform(5, 10, size=num_points)
yerr = np.random.normal(np.zeros(num_points), ysigma)
y = quadratic_model(a,b,c) + yerr

errorbar(x, y, yerr=ysigma, color="k", linestyle="none", marker="o", markersize=4)
```

```python
p_opt_lin, cov_lin, info, mesg, ier = optimize.leastsq(err_f(linear_model), x0=(0.,0.), args=(x, y, ysigma), full_output=True) 
print ier
```

```python
from scipy import linalg
```

```python
# Solve linear algebra least squares
Y = y
A = np.ones((len(Y),2))
A[:,1] = x
C = np.zeros((len(Y),len(Y)))
for ii in range(len(Y)):
    C[ii,ii] = ysigma[ii]

bestfit = np.dot(linalg.inv(C), Y.T)
bestfit = np.dot(A.T, bestfit)
bestfitvar = np.dot(linalg.inv(C), A)
bestfitvar = np.dot(A.T, bestfitvar)
bestfitvar = linalg.inv(bestfitvar)
bestfit = np.dot(bestfitvar, bestfit)
```

```python
b, m = bestfit
m_err, b_err = np.sqrt(bestfitvar[1,1]), np.sqrt(bestfitvar[0,0])
m_err, b_err
```

```python
def lnprob(model):
    return lambda p,x,y_data,y_err: -np.sum(((y_data - model(p, x))/y_err)**2)

nwalkers = 100
ndim = 2
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob(linear_model), args=[x,y,ysigma])
```

```python
p0 = np.random.rand(nwalkers*ndim).reshape((nwalkers,ndim))
pos, prob, state = sampler.run_mcmc(p0, 100)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 1000)
```

```python
# Get N samples from the posterior to plot
N = 100
samples = sampler.flatchain[np.random.randint(len(sampler.flatchain), size=N)]

# Compute uncertainty intervals over each parameter
ms = sampler.flatchain[:,0]
bs = sampler.flatchain[:,1]

display(Latex(r"Lev-Mar: $σ_m={0}$, $σ_b={1}$".format(np.sqrt(cov_lin[0,0]), np.sqrt(cov_lin[1,1]))))
display(Latex(r"Linear Algebra: $σ_m={0}$, $σ_b={1}$".format(m_err, b_err)))
display(Latex(r"MCMC: $σ_m={0}$, $σ_b={1}$".format(np.std(ms), np.std(bs))))
```
